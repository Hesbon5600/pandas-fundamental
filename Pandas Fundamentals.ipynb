```python
KEYS_TO_USE = ['id', 'all_artists', 'title', 'medium', 'acquisitionYear', 'height', 'width', 'dateText', 'units']
```

```python
import json
def get_records_from_file(file_path, keys_to_use):
    # Process a simgle json file and return a tupl econtaining specific fiels
    with open(file_path) as artwork_file:
        content = json.load(artwork_file)
        
        record = []
        for field in keys_to_use:
            record.append(content[field])
        
        return tuple(record)
```

```python
SAMPLE_JSON = os.path.join('collection-master', 'artworks', 'a', '000', 'a00001-1035.json')
```

```python
sample_record = get_records_from_file(SAMPLE_JSON, KEYS_TO_USE)
```

```python
pd.DataFrame(sample_record)
```

```python
def read_artworks_from_json(keys_to_use):
    """Traverse the directories with json files
    For the first file in each directory,
    call function for processing single file
    and go to the next directory"""
    JSON_ROOT = os.path.join('collection-master', 'artworks')
    artworks = []
    for root, _, files, in os.walk(JSON_ROOT):
        for f in files:
            if f.endswith('json'):
                record = get_records_from_file(os.path.join(root, f), keys_to_use)
                artworks.append(record)
            break
    df = pd.DataFrame.from_records(artworks, columns = keys_to_use, index='id')
    return df
```

```python
df = read_artworks_from_json(KEYS_TO_USE)
```

### Treating missing values

```python
# Fill missing values with the most frequent value in the series
def fill_values(series):
    values_counted = series.value_counts()
    if values_counted.empty:
        return series
    most_frequent = values_counted.index[0]
    new_medium = series.fillna(most_frequent)
    return new_medium
```

```python
# Update the dataframe
def transform_df(source_df):
    group_dfs = []
    for name, group_df in source_df.groupby('artist'):
        filled_df = group_df.copy()
        filled_df.loc[:, 'medium'] = fill_values(group_df['medium'])
        group_dfs.append(filled_df)
    new_df = pd.concat(group_dfs)
    return new_df
```

```python
# Now check the result
filled_df = transform_df(small_df)
```

```python
filled_df
```

### Built in methods

```python
# Transform
grouped_mediums = small_df.groupby('artist')['medium']
# .transform() method takes in a function as a parameter [The function
# must be on ethat takes in a series as a parameter and performs certain operations]
small_df.loc[:, 'medium'] = grouped_mediums.transform(fill_values)
```
